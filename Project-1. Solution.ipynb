3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# описываем функцию-преобразования для определения опыта работы соискателей в месяцах
def get_exp(x):
  try:
    if x.split()[3][0:3] == 'год' and x.split()[5][0:5] == 'месяц':
      exp = int(x.split()[2])*12 + int(x.split()[4])
    elif x.split()[3][0:3] == 'лет' and x.split()[5][0:5] == 'месяц':
      exp = int(x.split()[2])*12 + int(x.split()[4])
    elif x.split()[3][0:5] == 'месяц':
      exp = int(x.split()[2])
    else:
      exp = int(x.split()[2])*12
  except: exp = np.nan
  return exp
# формируем новый признак, содержащий информацию о том, сколько месяцев проработали соискатели
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(get_exp).astype(float).astype('Int64')
# удаляем признак "Опыт работы"
df = df.drop(['Опыт работы'], axis=1)
# рассчитываем и выводим медианный опыт работы соискателей
print('Медианный опыт работы соискателей составляет',df['Опыт работы (месяц)'].median (),'мес.')
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# описываем функцию-преобразования для определения категорий желаемой занятости соискателей
def encode_col(data, col):
  arr = pd.Series((', '.join(data[col].unique()).split(sep=', '))).unique().tolist()
  for i in arr:
    data[i]=data[col].apply(lambda x: i in x.split(sep=', '))
  return data
# создаем признаки-«мигалки» для каждой категории занятости, применяя функцию-преобразования к признакам "Занятость" и "График"
df = encode_col(df, 'Занятость')
df = encode_col(df, 'График')
# удаляем признаки "Занятость" и "График"
df = df.drop(['Занятость','График'], axis=1)
# определяем количество соискателей ищущих проектную работу и волонтёрство
print('Количество соискателей ищущих проектную работу и волонтёрство составляет',len(df[(df['волонтерство'] == True) & (df['проектная работа'] == True)]),'чел.')
print()
# определяем количество соискателей желающих работать вахтовым методом или с гибким графиком
print('Количество соискателей желающих работать вахтовым методом или с гибким графиком составляет',len(df[(df['вахтовый метод'] == True) & (df['гибкий график'] == True)]),'чел.')
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# считываем предоставленный массив курсов валют
ExchangeRates_path = '/content/drive/MyDrive/Colab Notebooks/ExchangeRates.csv'
df_curr = pd.read_csv(ExchangeRates_path, find_delimiter(ExchangeRates_path))
# приводим признак "date" из таблицы с валютами в формат datetime
df_curr['date']=pd.to_datetime(df_curr['date']).dt.date
# рассчитываем курс с учетом пропорции (признак 'close' приводим к величине курса за 1 единицу валюты)
df_curr['close']=df_curr['close']/df_curr['proportion']
# удаляем неиспользуемые признаки из массива курсов валют
df_curr = df_curr.drop(['proportion', 'per', 'time', 'vol'], axis=1)
# переводим признак "Обновление резюме" из таблицы с резюме в формат datetime и достаем из него дату
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме']).dt.date
# выделияем из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется
df[['ЗП', 'Валюта ЗП']] = df['ЗП'].str.split(" ", expand = True)
# преобразовываем тип данных столбца с желаемой заработной платой
df['ЗП'] = df['ЗП'].astype (float)
# определяем полный список валют, указанных в массиве резюме.
cv_codes = df['Валюта ЗП'].unique ()
# print(cv_codes) : (['руб.' 'бел.руб.' 'KZT' 'EUR' 'USD' 'грн.' 'сум' 'KGS' 'AZN'])
# определяем полный список валют, указанных в массиве курсов валют, дополнив его валютой "RUB"
iso_codes = ['RUB', 'BYN', 'KZT', 'EUR', 'USD', 'UAH', 'UZS', 'KGS', 'AZN']
# формируем словарь для переопределения наименований валют в соответствии со стандартом ISO
curr_codes=dict(zip(cv_codes, iso_codes))
# переопределяем наименования валют в соответствии со стандартом ISO
df['Валюта ЗП'] = df['Валюта ЗП'].map(curr_codes)
# присоединяем к таблице с резюме таблицу с курсами
df = df.merge(df_curr, left_on=['Обновление резюме','Валюта ЗП'], right_on=['date','currency'], how='left')
# заполняем пропуски признака close (для валюты "RUB") единицей 1 (курс рубля самого к себе)
df['close'] = df['close']. fillna (1)
# рассчитываем сумму зарплаты в рублях
df['ЗП (руб)'] = df['close']*df['ЗП']
# удаляем признак "ЗП" и промежуточные признаки "date", "currency" и "close"
df = df.drop(['ЗП', 'date','currency','close'], axis=1)
# определяем желаемую медианную заработную плату соискателей
print('Медианная желаемая заработная плата соискателей',(df['ЗП (руб)'].median()/1000).round(0),'тыс. руб.')
```
